# Tracé de courbes Paramétriques 2D

```python
from math import *
from pylab import *
from numpy import *

theta = linspace(0,2*pi,50)
x=cos(theta)
y=sin(theta)                        # on a la représentation paramétrique d'un cercle dans le plan

plot(x,y,'b')
axis('equal')                       # instruction très importante car sinon le cercle apparaît comme une ellipse !
grid()
title('tracé dun cercle trigonométrique')
show
```

```python
from math import *
from pylab import *
from mpl_toolkits.mplot3d import *

ax=Axes3D(figure())

def selledecheval(x,y):
    return x**2-y**2

x=linspace(-2,2,200)
y=linspace(-2,2,200)
x,y=meshgrid(x,y)
z=selledecheval(x,y)

ax.plot_surface(x,y,z)
xticks([-2,-1,0,1,2])
yticks([-2,-1,0,1,2])

show()
```

```python
from math import *
from pylab import *
from mpl_toolkits.mplot3d import *

ax=Axes3D(figure())

u, v = mgrid[0:2*pi:500j, 0:pi:400j]
x = cos(u)*sin(v)
y = sin(u)*sin(v)
z = cos(v)

ax.plot_surface(x, y, z, color="r")
show()
```

```python
from math import *
from pylab import *
from mpl_toolkits.mplot3d import *

ax=Axes3D(figure())

def surface3D_1(x,y):
    u=x**2+y**2
    v=sqrt(u)
    w=sin(v)
    return w

x=linspace(-5,5,100)
y=linspace(-5,5,100)
x,y=meshgrid(x,y)
z=surface3D_1(x,y)

ax.plot_surface(x,y,z)

show()
```

```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z,cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
