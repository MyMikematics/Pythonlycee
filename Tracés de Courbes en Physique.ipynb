```python
from math import *
from pylab import *

figure(figsize=(12,9), dpi=80)

def U(t,Q):
    return exp(-2/Q*t)*cos(2*pi*t)             # Fonction oscillation amortie U dépendant d'un paramètre Q

t = linspace(0,10,400)  

plot(t,U(t,2),'r')
plot(t,U(t,10),'b')

xlim(0, 8)
ylim(-1,1)
legend(('t (s)','U (V)'),loc='upper right') 
show()
```

```python
from math import *
from pylab import *

x = logspace(-2,2, 500)                         # x=omega/omega0 = Abscisses

def H(x, Q):                                    # Fonction de transfert. j ?
    return 1/(1+1j*x/Q-x**2)

Qvaleurs = [0.1, 1/sqrt(2), 3]                  # Liste des Valeurs désirées du facteur de qualité

for Q in Qvaleurs :                       
       plot(log10(x), 20*log10(abs(H(x,Q))), 180/pi * angle(H(x,Q)))

xlim(-2,2,100)
ylim(-180,0)
show()
        
```
